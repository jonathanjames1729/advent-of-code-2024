```python
import itertools

import advent_of_code

TEST=False

NUMERIC = {
    '0': (1, 0),
    '1': (0, 1),
    '2': (1, 1),
    '3': (2, 1),
    '4': (0, 2),
    '5': (1, 2),
    '6': (2, 2),
    '7': (0, 3),
    '8': (1, 3),
    '9': (2, 3),
    'A': (2, 0),
    '':  (0, 0)
}

DIRECTIONAL = {
    '<': (0, 0),
    'v': (1, 0),
    '>': (2, 0),
    '^': (1, 1),
    'A': (2, 1),
    '':  (0, 1)
}

def f(list_of_options):
    counts = list(map(len, list_of_options))
    indexes = [0] * len(counts)
    while True:
        result = []
        for i, index in enumerate(indexes):
            result.extend(list_of_options[i][index])
        yield result
        i = 0
        for i, count in enumerate(counts):
            indexes[i] += 1
            if indexes[i] < count:
                break
            indexes[i] = 0
        else:
            break

def count_bits(n):
    count = 0
    while n:
        count += 1
        n = n & (n - 1)
    return count

def get_directions(buttons, keypad):
    positions = ['A']
    positions.extend(buttons)
    avoid_x, avoid_y = keypad['']
    result = []
    for one, two in itertools.pairwise(positions):
        one_x, one_y = keypad[one]
        two_x, two_y = keypad[two]
        horizontal, width = ('>', two_x - one_x) if two_x > one_x else ('<', one_x - two_x)
        vertical, height = ('^', two_y - one_y) if two_y > one_y else ('v', one_y - two_y)
        options = []
        for n in range(2 ** width - 1, 2 ** (width + height)):
            if count_bits(n) == width:
                option = [horizontal if n & (2 ** i) else vertical for i in range(width + height)]
                x, y = (one_x, one_y)
                for direction in option:
                    if direction == '>':
                        x += 1
                    elif direction == '<':
                        x -= 1
                    elif direction == '^':
                        y += 1
                    else:
                        y -= 1
                    if x == avoid_x and y == avoid_y:
                        break
                else:
                    option.append('A')
                    options.append(option)
        result.append(options)
    return result

result = 0
for line in advent_of_code.load_data('day21.txt', test=TEST):
    value = int(line[:-1])
    min_count = None
    min_buttons = None
    list_of_options = get_directions(line, NUMERIC)
    for buttons in f(list_of_options):
        list_of_options2 = get_directions(buttons, DIRECTIONAL)
        for buttons2 in f(list_of_options2):
            list_of_options3 = get_directions(buttons2, DIRECTIONAL)
            for buttons3 in f(list_of_options3):
                count = len(buttons3)
                if not min_count or count < min_count:
                    min_count = count
                    min_buttons = buttons3
    print(line, value, min_count, ''.join(min_buttons))
    result += value * min_count

print(result)
```

```python
def get_directions_count(buttons, keypad):
    positions = ['A']
    positions.extend(buttons)
    result = 0
    for one, two in itertools.pairwise(positions):
        one_x, one_y = keypad[one]
        two_x, two_y = keypad[two]
        result += abs(two_x - one_x) + abs(two_y - one_y) + 1
    return result

result = 0
for line in advent_of_code.load_data('day21.txt', test=TEST):
    value = int(line[:-1])
    min_count = None
    min_buttons = None
    list_of_options = get_directions(line, NUMERIC)
    for buttons in f(list_of_options):
        list_of_options2 = get_directions(buttons, DIRECTIONAL)
        for buttons2 in f(list_of_options2):
            count = get_directions_count(buttons2, DIRECTIONAL)
            if not min_count or count < min_count:
                min_count = count
    print(line, value, min_count)
    result += value * min_count

print(result)
```
