```python
import advent_of_code

TEST = False
SIZE = 10 if TEST else 50
DIRECTIONS = {
    '>': (1, 0),
    'v': (0, 1),
    '<': (-1, 0),
    '^': (0, -1)
}

movements = []
robot_x, robot_y = (None, None)
walls = set()
boxes = set()

y = 0
for line in advent_of_code.load_data('day15.txt', test=TEST):
    if not line:
        y = None
    elif y is None:
        movements.extend(map(DIRECTIONS.get, line))
    else:
        for x, space in enumerate(line):
            if space == '#':
                walls.add((x, y))
            elif space == 'O':
                boxes.add((x, y))
            elif space == '@':
                robot_x, robot_y = (x, y)
        y += 1

for dx, dy in movements:
    x, y = (robot_x, robot_y)
    pushing = False
    can_move = None
    while can_move is None:
        x += dx
        y += dy
        if (x, y) in walls:
            can_move = False
        elif (x, y) in boxes:
            pushing = True
        else:
            can_move = True
    if can_move:
        robot_x += dx
        robot_y += dy
        if pushing:
            boxes.remove((robot_x, robot_y))
            boxes.add((x, y))

print(sum(x + (100 * y) for x, y in boxes))

for y in range(SIZE):
    line = []
    for x in range(SIZE):
        if (x, y) in walls:
            line.append('#')
        elif (x, y) in boxes:
            line.append('O')
        elif x == robot_x and y == robot_y:
            line.append('@')
        else:
            line.append('.')
    print(''.join(line))
```

```python
movements = []
robot_x, robot_y = (None, None)
walls = set()
boxes = []

y = 0
for line in advent_of_code.load_data('day15.txt', test=TEST):
    if not line:
        y = None
    elif y is None:
        movements.extend(map(DIRECTIONS.get, line))
    else:
        for x, space in enumerate(line):
            if space == '#':
                walls.add((2 * x, y))
                walls.add(((2 * x) + 1, y))
            elif space == 'O':
                boxes.append([(2 * x, y), ((2 * x) + 1, y)])
            elif space == '@':
                robot_x, robot_y = (2 * x, y)
        y += 1

for dx, dy in movements:
    moving = []
    moved = set()
    moved.add((robot_x + dx, robot_y + dy))
    while moved:
        if moved.intersection(walls):
            moved = None
        else:
            new_moved = set()
            for i, box in enumerate(boxes):
                if i not in moving and moved.intersection(box):
                    moving.append(i)
                    for x, y in box:
                        new_moved.add((x +dx, y + dy))
            moved = new_moved
    if moved is not None:
        robot_x += dx
        robot_y += dy
        for i in moving:
            boxes[i] = [(x + dx, y + dy) for x, y in boxes[i]]

print(sum(box[0][0] + (100 * box[0][1]) for box in boxes))

lines = [['.'] * (SIZE * 2) for _ in range(SIZE)]
for x, y in walls:
    lines[y][x] = '#'
for box in boxes:
    x, y = box[0]
    lines[y][x] = '['
    x, y = box[1]
    lines[y][x] = ']'
lines[robot_y][robot_x] = '@'
for line in lines:
    print(''.join(line))
```
