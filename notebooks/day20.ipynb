```python
import advent_of_code

TEST = False
SIZE = 15 if TEST else 141

start = None
end = None
walls = set()
for y, line in enumerate(advent_of_code.load_data('day20.txt', test = TEST)):
    for x, space in enumerate(line):
        if space == '#':
            walls.add((x, y))
        elif space == 'S':
            start = (x, y)
        elif space == 'E':
            end = (x, y)

path = [None, start]
while path[-1] != end:
    x, y = path[-1]
    for dx, dy in ((1, 0), (0, 1), (-1, 0), (0, -1)):
        position = (x + dx, y + dy)
        if position not in walls and position != path[-2]:
            path.append(position)
            break
path.pop(0)

print(f'Course time: {len(path) - 1}')

cheats = {}
for x, y in walls:
    if x == 0 or y == 0 or x == (SIZE -1) or y == (SIZE - 1):
        continue
    distances = []
    for dx, dy in ((1, 0), (0, 1), (-1, 0), (0, -1)):
        position = (x + dx, y + dy)
        if position not in walls:
            try:
                distance = path.index(position)
            except ValueError:
                continue
            distances.append(distance)
    distances.sort()
    if len(distances) == 2:
        d = distances[1] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
    elif len(distances) == 3:
        d = distances[1] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[2] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[2] - distances[1] - 2
        cheats[d] = cheats.get(d, 0) + 1
    elif len(distances) == 4:
        d = distances[1] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[2] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[3] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[2] - distances[1] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[3] - distances[1] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[3] - distances[2] - 2
        cheats[d] = cheats.get(d, 0) + 1

for distance in sorted(cheats.keys()):
    if distance == 0:
        continue
    print(f'{cheats[distance]} cheat(s) saving {distance} picoseconds')

if not TEST:
    total = 0
    for distance, count in cheats.items():
        if distance >= 100:
            total += count
    print(f'{total} cheats save at least 100 picoseconds')
```
