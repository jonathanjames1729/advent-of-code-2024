```python
import advent_of_code

TEST = False
SIZE = 15 if TEST else 141

start = None
end = None
walls = set()
for y, line in enumerate(advent_of_code.load_data('day20.txt', test = TEST)):
    for x, space in enumerate(line):
        if space == '#':
            walls.add((x, y))
        elif space == 'S':
            start = (x, y)
        elif space == 'E':
            end = (x, y)

path = [None, start]
while path[-1] != end:
    x, y = path[-1]
    for dx, dy in ((1, 0), (0, 1), (-1, 0), (0, -1)):
        position = (x + dx, y + dy)
        if position not in walls and position != path[-2]:
            path.append(position)
            break
path.pop(0)

print(f'Course time: {len(path) - 1}')

cheats = {}
for x, y in walls:
    if x == 0 or y == 0 or x == (SIZE -1) or y == (SIZE - 1):
        continue
    distances = []
    for dx, dy in ((1, 0), (0, 1), (-1, 0), (0, -1)):
        position = (x + dx, y + dy)
        if position not in walls:
            try:
                distance = path.index(position)
            except ValueError:
                continue
            distances.append(distance)
    distances.sort()
    if len(distances) == 2:
        d = distances[1] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
    elif len(distances) == 3:
        d = distances[1] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[2] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[2] - distances[1] - 2
        cheats[d] = cheats.get(d, 0) + 1
    elif len(distances) == 4:
        d = distances[1] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[2] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[3] - distances[0] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[2] - distances[1] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[3] - distances[1] - 2
        cheats[d] = cheats.get(d, 0) + 1
        d = distances[3] - distances[2] - 2
        cheats[d] = cheats.get(d, 0) + 1

for distance in sorted(cheats.keys()):
    if distance == 0:
        continue
    print(f'{cheats[distance]} cheat(s) saving {distance} picoseconds')

if not TEST:
    total = 0
    for distance, count in cheats.items():
        if distance >= 100:
            total += count
    print(f'{total} cheats save at least 100 picoseconds')
```

```python
MIN_SAVING = 50 if TEST else 100

count = 0
cheats = {}
for cost, (x, y) in enumerate(path):
    savings = {}
    visited = set()
    process = set()
    for dx, dy in ((1, 0), (0, 1), (-1, 0), (0, -1)):
        next_x = x + dx
        next_y = y + dy
        if next_x in (-1, SIZE) or next_y in (-1, SIZE):
            continue
        process.add((next_x, next_y))
    i = 2
    while process and i <= 20:
        next_process = set()
        for current_x, current_y in process:
            for dx, dy in ((1, 0), (0, 1), (-1, 0), (0, -1)):
                next_x = current_x + dx
                next_y = current_y + dy
                next_position = (next_x, next_y)
                if next_x in (-1, SIZE) or next_y in (-1, SIZE) or next_position in visited:
                    continue
                if next_position not in walls:
                    try:
                        current_cost = path.index(next_position)
                    except ValueError:
                        continue
                    next_cost = cost + i
                    if next_cost < current_cost:
                        savings[next_position] = max(savings.get(next_position, 0), current_cost - next_cost)
                next_process.add(next_position)    
        visited.update(process)
        process = next_process
        i += 1
    for saving in savings.values():
        if saving >= MIN_SAVING:
            cheats[saving] = cheats.get(saving, 0) + 1
            count += 1

for distance in sorted(cheats.keys()):
    print(f'{cheats[distance]} cheat(s) saving {distance} picoseconds')
print(count)
```
